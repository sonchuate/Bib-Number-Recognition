Training

```python
import yaml

# Dữ liệu bạn muốn ghi vào file YAML
data = {
    'train': '',    #lưu vị trí tập train
    'val': '',      #lưu vị trí tập val
    'nc': '1',      #num class
    'name': ['bib number']
}

# Ghi dữ liệu vào file YAML
with open('config.yaml', 'w') as file:
    yaml.dump(data, file)
```

```python
model = YOLO('yolov8s.pt')
model.train(data='config.yaml', epochs=1)
```
